### Droping some unnecessary column which are not relevant that much

```python
df_covid.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1)
```

```python
df_covid.head()
```

### Convert Date column data to datetime format

```python
df_covid['Date']=pd.to_datetime(df_covid['Date'],format='%Y-%m-%d')
```

```python
df_covid.head()
```

### Create a column Active Cases to show active cases

```python
df_covid['ACtive_cases']=df_covid['Confirmed']-(df_covid['Cured']+df_covid['Deaths'])
```

```python
df_covid.head()
```

```python
df_covid.tail()
```

## pivot table using pandas

```python
statewise=pd.pivot_table(df_covid,values=['Confirmed','Deaths','Cured'],
                         index='State/UnionTerritory',aggfunc='max')
```

```python
statewise.sort_values('Confirmed',ascending=False)
```

### Add 2 columns Recovery rate and Mortality rate statewise by some calculation

```python
statewise['RecoveryRate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['MortalityRate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise
```

```python
statewise=statewise.sort_values('Confirmed',ascending=False)
```

```python
statewise
```

### Visualize data with cmap that will give better understanding for data.

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
# Top 10 Active Cases State

Top_10=df_covid.groupby(by='State/UnionTerritory').max()[['ACtive_cases','Date']].sort_values(['ACtive_cases'],ascending=False).reset_index()
```

```python
plt.figure(figsize=(12,7))
plt.title('Top 10 State with actice cases')
ax=sns.barplot(data=Top_10.iloc[:10],y='ACtive_cases',x='State/UnionTerritory',linewidth=2)
```

### Top 10 state with respect to deaths.

```python
top_10_deaths=statewise['Deaths'].sort_values(ascending=False).head(12)
```

```python
top_10_deaths
```

### Covid cases trends in top covid effected states

```python
plt.figure(figsize=(12,6))
ax=sns.lineplot(data=df_covid[df_covid['State/UnionTerritory'].isin(['Maharastra','Karnataka','Tamil Nadu','Delhi','Uttar Pradesh'])],x='Date',y='ACtive_cases',hue='State/UnionTerritory')
plt.title('Top 5 Affected State in Inda')
```

```python
vaccine=df_vacc.drop(columns=['Sputnik V (Doses Administered)','18-44 Years (Doses Administered)','45-60 Years (Doses Administered)','60+ Years (Doses Administered)','AEFI'],axis=1)
```

```python
plt.pie([male,female],labels=['Male','Female'],autopct='%.2f%%')
plt.title('MAle and Female vacination')
```

```python
# Remove rows where state=India

vaccine=df_vacc[df_vacc.State!='India']
```

```python
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
```

```python
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python
max_vacc=pd.DataFrame(max_vac)
```

```python
max_vacc
```

```python
sns.barplot(data=max_vac,x=max_vac.index,y=max_vac.Total)
```

```python
min_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac=min_vac.sort_values('Total',ascending=True)[:5]
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=min_vac,x=min_vac.index,y=min_vac.Total)
plt.xticks(rotation=90)
```
